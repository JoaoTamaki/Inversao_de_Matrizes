```python
import pandas as pd
import glob
import os
import numpy as np
import matplotlib.pyplot as plt 
```

```python
files = []
for f in glob.glob(os.path.join("Outputs", "*.log")):
    nome_arq = f.split('\\')
    nome_arq = nome_arq[1]
    nome_arq = nome_arq.split('.')
    nome_arq = nome_arq[0]
    parametros = nome_arq.split('_') 
    files.append((int(parametros[1]), f))
files = sorted(files)
```

```python
FLOPS_T1_OP1 =[]
FLOPS_T1_OP2 =[]
FLOPS_T2_OP1 =[]
FLOPS_T2_OP2 =[]

L3_T1_OP1 =[]
L3_T1_OP2 =[]
L3_T2_OP1 =[]
L3_T2_OP2 =[]

L2CACHE_T1_OP1 =[]
L2CACHE_T1_OP2 =[]
L2CACHE_T2_OP1 =[]
L2CACHE_T2_OP2 =[]

TEMPO_T1_OP1 =[]
TEMPO_T1_OP2 =[]
TEMPO_T2_OP1 =[]
TEMPO_T2_OP2 =[]
 
N = []

for f in files:
    (aux, txt)  = f
    arquivo = open(txt, "r")
    nome_arq = txt.split('\\')
    nome_arq = nome_arq[1]
    nome_arq = nome_arq.split('.')
    nome_arq = nome_arq[0]
    parametros = nome_arq.split('_') 
    content = arquivo.readlines()
    if (parametros[2] == 'FLOPS'):
        if(parametros[4] == 'T1'):
            if(parametros[0] == 'OP1'):
                aux = content[37].split(' ')
                FLOPS_T1_OP1.append(float(aux[len(aux)-2]))
                aux = content[15].split(' ')
                aux = aux[len(aux)-1].split('\n')
                N.append(int(parametros[1]))
                TEMPO_T1_OP1.append(float(aux[0]))
            if(parametros[0] == 'OP2'):
                aux = content[37].split(' ')
                FLOPS_T1_OP2.append(float(aux[len(aux)-2]))
                aux = content[15].split(' ')
                aux = aux[len(aux)-1].split('\n')
                TEMPO_T1_OP2.append(float(aux[0]))
            
        if(parametros[4] == 'T2'):
            if(parametros[0] == 'OP1'):
                aux = content[38].split(' ')
                FLOPS_T2_OP1.append(float(aux[len(aux)-2]))
                aux = content[16].split(' ')
                aux = aux[len(aux)-1].split('\n')
                TEMPO_T2_OP1.append(float(aux[0]))
            if(parametros[0] == 'OP2'):
                aux = content[38].split(' ')
                FLOPS_T2_OP2.append(float(aux[len(aux)-2]))
                aux = content[16].split(' ')
                aux = aux[len(aux)-1].split('\n')
                TEMPO_T2_OP2.append(float(aux[0]))

    if (parametros[2] == 'L2CACHE'):
        if(parametros[3] == 'T1'):
            if(parametros[0] == 'OP1'):
                aux = content[38].split(' ')
                L2CACHE_T1_OP1.append(float(aux[len(aux)-2]))
            if(parametros[0] == 'OP2'):
                aux = content[38].split(' ')
                L2CACHE_T1_OP2.append(float(aux[len(aux)-2]))
            
        if(parametros[3] == 'T2'):
            if(parametros[0] == 'OP1'):
                aux = content[39].split(' ')
                L2CACHE_T2_OP1.append(float(aux[len(aux)-2]))
            if(parametros[0] == 'OP2'):
                aux = content[39].split(' ')
                L2CACHE_T2_OP2.append(float(aux[len(aux)-2]))

    if (parametros[2] == 'L3'):
        if(parametros[3] == 'T1'):
            if(parametros[0] == 'OP1'):
                aux = content[40].split(' ')
                L3_T1_OP1.append(float(aux[len(aux)-2]))
            if(parametros[0] == 'OP2'):
                aux = content[40].split(' ')
                L3_T1_OP2.append(float(aux[len(aux)-2]))
            
        if(parametros[3] == 'T2'):
            if(parametros[0] == 'OP1'):
                aux = content[41].split(' ')
                L3_T2_OP1.append(float(aux[len(aux)-2]))
            if(parametros[0] == 'OP2'):
                aux = content[41].split(' ')
                L3_T2_OP2.append(float(aux[len(aux)-2]))
                
    arquivo.close
```

```python
for i in L2CACHE_T1_OP1:
    print(i)
```

```python
fig = plt.figure(figsize=(8, 6))

plt.scatter(N, L3_T1_OP1, label="T1_op1")
plt.scatter(N, L3_T1_OP2, label="T1_op2")
plt.scatter(N, L3_T2_OP1, label="T2_op1")
plt.scatter(N, L3_T2_OP2, label="T2_op2")

plt.plot(N, L3_T1_OP1, label="T1_op1")
plt.plot(N, L3_T1_OP2, label="T1_op2")
plt.plot(N, L3_T2_OP1, label="T2_op1")
plt.plot(N, L3_T2_OP2, label="T2_op2")

plt.yscale("log")
plt.title("L3",fontsize=25)
plt.xlabel("N")
plt.ylabel("Memory bandwidth [MBytes/s]")
plt.legend()
plt.show()
```
